```

import mlflow
import mlflow.sklearn

model_path = "isolation_forest_model.pkl"
scaler_path = "scaler.pkl"

mlflow.set_tracking_uri("http://mlflow-service.<your-project>.svc.cluster.local:5000")
mlflow.set_experiment("fraude-smart-meters")

with mlflow.start_run():
    mlflow.sklearn.log_model(artifact_path="model", sk_model=mlflow.sklearn.load_model(model_path))
    mlflow.log_artifact(scaler_path)
    mlflow.log_param("model_type", "IsolationForest")
    mlflow.log_param("contamination", 0.02)
```
